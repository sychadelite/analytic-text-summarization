```python
# https://ksnugroho.medium.com/dasar-text-preprocessing-dengan-python-a4fa52608ffe

# 1. Melakukan Case folding (mengubah semua huruf menjadi huruf kecil dan menghilangkan newline /n).
paperText = open('C:/Users/Buzz Lightyear/Downloads/Evaluasi Sentence Extraction pada Peringkasan Dokumen Otomatis.txt','r', encoding='utf-8')
paperText_lowercase_strip = [line.lower().rstrip() for line in paperText]
hasilText = paperText_lowercase_strip

print(hasilText)
```

```python
# 2. Menghapus angka menggunakan RE (Regular Expression)
import re

hasilText_numbers = hasilText
hasilText_removed_numbers = re.sub(r"\d+", "", str(hasilText_numbers))

print(hasilText_removed_numbers)
```

```python
# 3. Menghapus tanda baca
import string

hasilText_notation = str(hasilText_removed_numbers)
hasilText_removed_notations = hasilText_notation.translate(str.maketrans("", "", string.punctuation))

print(hasilText_removed_notations)
```

```python
# 4. Tokenisasi kata (NLTK)
import nltk
from nltk.tokenize import word_tokenize
 
list_of_token = str(hasilText_removed_notations)
hasilText_tokens = nltk.tokenize.word_tokenize(list_of_token)

print(hasilText_tokens)
```

```python
# 5. Menghapus stop words bahasa indonesia dari token
from nltk.corpus import stopwords

listStopword =  set(stopwords.words('indonesian'))
 
removed = []
for t in hasilText_tokens:
    if t not in listStopword:
        removed.append(t)

hasilTextToken = removed
print(hasilTextToken)
```

```python
# 6. Frekuensi kemunculan setiap token
from nltk.probability import FreqDist

freq = hasilTextToken 

freq_hasilTextToken = nltk.FreqDist(freq)
print(freq_hasilTextToken.most_common())
```

```python
# https://stackoverflow.com/questions/37427673/sorting-freqdist-in-nltk-with-get-vs-get
# https://realpython.com/iterate-through-dictionary-python/

# 7. Parse array Frequency
love = freq_hasilTextToken
asd = love.most_common()

dsa = dict(love)
type(dsa)

dsa_sort = sorted(dsa.items(), key = lambda kv:(kv[1], kv[0]), reverse=True)
# print(dsa_sort)

fuck = dict(dsa_sort)

for key in fuck:
 print(key, ':', fuck[key])
```

```python
# 8. Visualisasi dengan WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud

dictionary = Counter(freq_hasilTextToken)

cloud = WordCloud(max_font_size=80,colormap="hsv").generate_from_frequencies(dictionary)
plt.figure(figsize=(16,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
